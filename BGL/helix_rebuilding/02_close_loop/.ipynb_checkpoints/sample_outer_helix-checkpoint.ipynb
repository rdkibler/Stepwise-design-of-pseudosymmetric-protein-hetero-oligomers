```python
calculator = pyrosetta.rosetta.protocols.helical_bundle.BundleParametrizationCalculator()
calculator.init_from_file("alpha_helix_100.crick_params")
calculator.set_use_degrees(True)

def genhelix(d_omega0=0, d_omega1=0, invert=True, r0=0, z1_offset=0, hlen = 15):
    pose = pyrosetta.pose_from_sequence("A" * hlen)
    
    '''
    1 r0
    2 omega0
    3 delta_omega0
    4 delta_omega1
    5 delta_t
    6 z0_offset
    7 z1_offset
    8 epsilon
    9 residues_per_repeat
    10 repeating_unit_offset
    11 atoms_per_residue
    12 r1_peratom
    13 omega1
    14 z1
    15 delta_omega1_peratom
    16 delta_z1_peratom
    17 invert
    18 set_dihedrals
    19 set_bondangles
    20 set_bondlengths
    '''
    
    calculator.parameter(1).set_value(r0)
    calculator.parameter(3).set_value(d_omega0)
    calculator.parameter(4).set_value(d_omega1)
    calculator.parameter(7).set_value(z1_offset + (1.4 * invert))
    calculator.parameter(17).set_value(invert)
    
    calculator.build_helix(pose,1,hlen)
    return pose

for i in range(1,20):
    pyrosetta.dump_pdb(genhelix(0,0,False,0,0,),f"h{i}.pdb")
#pwhen inverting helix, use z1_offset=1.4
```

```python
#target generation

import numpy as np

n_anchor_targets = []
c_anchor_targets = []

z1_offset = 0.0
d_omega0 = 0
invert = False
r0 = 5.3

for d_omega1 in np.arange(0, 360, 30):
    n_anchor_targets.append(genhelix(d_omega0, d_omega1, invert, r0, z1_offset))

d_omega0 = 180
invert = True

for d_omega1 in np.arange(0, 360, 30):
    c_anchor_targets.append(genhelix(d_omega0, d_omega1, invert, r0, z1_offset))
```

```python
pose = pyrosetta.pose_from_sequence("AAAA")

def get_ids_and_coords(pose):
    ids = pyrosetta.rosetta.utility.vector1_core_id_AtomID()
    for i,res in enumerate(pose.residues):
        ros_i = i+1
        for j in range(1,res.natoms()+1):
            ids.append(pyrosetta.AtomID(j,ros_i))
    coords = pyrosetta.rosetta.utility.vector1_numeric_xyzVector_double_t()
    pose.batch_get_xyz(ids,coords) 
    return ids, coords

#use pose.batch_set_xyz(ids,coords) to set new coords
ids, coords = get_ids_and_coords(pose)

def vector1_numeric_xyzVector_double_t__to__numpy_arr(ros_vec):
    return np.array([xyz for xyz in ros_vec])

def numpy_arr__to__vector1_numeric_xyzVector_double_t(np_arr):
    coords = pyrosetta.rosetta.utility.vector1_numeric_xyzVector_double_t()
    for xyz in np_arr:
        x,y,z = xyz
        coords.append(pyrosetta.rosetta.numeric.xyzVector_double_t(x_a=x,y_a=y,z_a=z))
    return coords

assert(coords == numpy_arr__to__vector1_numeric_xyzVector_double_t(vector1_numeric_xyzVector_double_t__to__numpy_arr(coords)))
```

```python
def find_termius_center(pose, terminus):
    xs = []
    ys = []
    zs = []
    for res in pose.residues:
        x,y,z = res.atom('CA').xyz()
        xs.append(x)
        ys.append(y)
        zs.append(z)
        
    avg_x = sum(xs)/len(xs)
    avg_y = sum(ys)/len(ys)
    avg_z = sum(zs)/len(zs)
    if terminus in "nN":
        center = np.array([avg_x, avg_y, pose.residues[1].atom('CA').xyz()[2]])
    else:
        center = np.array([avg_x, avg_y, pose.residues[len(pose.residues)].atom('CA').xyz()[2]])
    return center

def random_tilt_u(sigma=0.5):
    a = np.append(np.random.normal(scale = sigma, size=2), [3.0])
    b = np.linalg.norm(a)
    u = a/b #unit vector for a random rotation
    return u


##HOW?
def random_tilt_R(sigma=0.5):
    a = np.array([0,0,1])
    b = random_tilt_u(sigma)
    
    v = np.cross(a,b)
    s = np.linalg.norm(v) #?
    c = a.dot(b)
    I = np.identity(3)
    
    #vXStr = '{} {} {}; {} {} {}; {} {} {}'.format(0, -v[2], v[1], v[2], 0, -v[0], -v[1], v[0], 0)
    #k = np.matrix(vXStr)
    
    k = np.array([[0 , -v[2], v[1]],[v[2], 0, -v[0]],[-v[1], v[0], 0]])
    
    R = I + k + np.matmul(k,k) * ((1 -c)/(s**2))
    return R


pose = c_anchor_targets[0]
print(find_termius_center(pose, "c"))
pose = n_anchor_targets[0]
print(find_termius_center(pose, "n"))
print(random_tilt_u())
print(random_tilt_u())
print(random_tilt_u())
print(random_tilt_u())
print(random_tilt_R())
```

```python
#target perturbation


#this func will take a straight helix and randomly swing everything around a point at the top center of the helix
def random_swing(pose, terminus, sigma=0.5, copies = 1):
    
    ids, coords = get_ids_and_coords(pose)
    np_coords = vector1_numeric_xyzVector_double_t__to__numpy_arr(coords)
    center = find_termius_center(pose, terminus)
    centered_coords = np_coords - center
    
    #print(centered_coords)
    
    output_poses = []
    
    for i in range(copies):
        R = random_tilt_R()

        tilted_coords = R.dot(centered_coords.T).T
        
        #return to center
        tilted_coords = tilted_coords + center
        tilted_coords = numpy_arr__to__vector1_numeric_xyzVector_double_t(tilted_coords)
        
        tilted_pose = pyrosetta.Pose()
        tilted_pose.detached_copy(pose)
        
        tilted_pose.batch_set_xyz(ids,tilted_coords)
        
        output_poses.append(tilted_pose)
    return output_poses
```

```python
def clash_freq(pose, targets):
    #pose is already aligned to target. All I need to do is run though them and see if they collide
    scorefxn = pyrosetta.ScoreFunction() 
    scorefxn.set_weight(pyrosetta.rosetta.core.scoring.fa_rep, 1.0) 
    pose_sc = scorefxn(pose)
    clash_check_pose = pyrosetta.Pose()
    
    num_clash = 0
    num_total = 0
    
    for i, target in enumerate(targets):
        num_total += 1
        
        target_sc = scorefxn(target)
        clash_check_pose.detached_copy(pose)
        clash_check_pose.append_pose_by_jump(target,1)
        clash_check_sc = scorefxn(clash_check_pose)
        diff = clash_check_sc - (target_sc + pose_sc)

        if abs(diff) > 500:
            num_clash += 1

    return float(num_clash)/float(num_total)
```

```python
n_super_aligned_anchors = []
c_super_aligned_anchors = []

superimposer = pyrosetta.rosetta.protocols.simple_moves.SuperimposeMover()
superimposer.set_ca_only(True)

for i, n_tgt in enumerate(n_anchor_targets):
    superimposer.set_reference_pose(n_tgt)
    for j, anchor in enumerate(n_term_anchors):
        
        aligned_pose = pyrosetta.Pose()
        aligned_pose.detached_copy(anchor)
        superimposer.set_target_range(start=1, end=15)
        superimposer.apply(aligned_pose)
        bad_score = clash_freq(aligned_pose, c_anchor_targets)
        
        print("n",i,j, bad_score)       
        
        if bad_score < 0.7:
            n_super_aligned_anchors.append(aligned_pose)

print(len(n_super_aligned_anchors))
pickle.dump( n_super_aligned_anchors, open( "nsaa.p", "wb" ) )
import random
for i,n in enumerate(random.sample(n_super_aligned_anchors, 20)):
    pyrosetta.dump_pdb(n, f"n{i}.pdb")   
            
for i, c_tgt in enumerate(c_anchor_targets):
    superimposer.set_reference_pose(c_tgt)
    for j, anchor in enumerate(c_term_anchors):
        
        aligned_pose = pyrosetta.Pose()
        aligned_pose.detached_copy(anchor)
        superimposer.set_target_range(start=len(anchor.residues)-14, end=len(anchor.residues))
        superimposer.apply(aligned_pose)
        bad_score = clash_freq(aligned_pose, n_anchor_targets)
        
        print("c",i,j, bad_score)
        
        if bad_score < 0.7:
            c_super_aligned_anchors.append(aligned_pose)
            
print(len(c_super_aligned_anchors))
pickle.dump( c_super_aligned_anchors, open( "csaa.p", "wb" ) )
for i,c in enumerate(random.sample(c_super_aligned_anchors, 20)):
    pyrosetta.dump_pdb(c, f"c{i}.pdb")
```
